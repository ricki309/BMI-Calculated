```python
while True:
    nama = input("Masukkan Nama Anda: ").strip()
    if nama != "":
        break
    else:
        print("Nama tidak boleh kosong. Silakan coba lagi.")

while True:
    print("Pilih Jenis Kelamin:")
    print("1. Laki-laki")
    print("2. Perempuan")

    pilihan_jenis_kelamin = input("Masukkan pilihan (1 atau 2): ").strip()

    if pilihan_jenis_kelamin == "1":
        jenis_kelamin = "Laki-laki"
        break
    elif pilihan_jenis_kelamin == "2":
        jenis_kelamin = "Perempuan"
        break
    else:
        print("Jenis_kelamin tidak boleh kosong. Silakan coba lagi.")

# Validasi tinggi badan
while True:
    tinggi_input = input("Masukkan tinggi badan dalam satuan cm: ").strip()
    if tinggi_input.isdigit():
        tinggi_badan = int(tinggi_input)
        if tinggi_badan > 0:
            break
        else:
            print("⚠️ Tinggi badan tidak boleh nol atau negatif!")
    else:
        print("⚠️ Masukkan hanya angka (contoh: 170)!")

# Validasi berat badan
while True:
    berat_input = input("Masukkan berat badan dalam satuan kg: ").strip()
    if berat_input.isdigit():
        berat_badan = int(berat_input)
        if berat_badan > 0:
            break
        else:
            print("⚠️ Berat badan tidak boleh nol atau negatif!")
    else:
        print("⚠️ Masukkan hanya angka (contoh: 60)!")

tinggi_badan_meter = 1/100 * tinggi_badan

tinggi_badan_kuadrat =  tinggi_badan_meter * tinggi_badan_meter

BMI = berat_badan / tinggi_badan_kuadrat

BMI_format = round(BMI, 1)

if BMI_format > 0:
    if BMI_format < 18.5:
        status = "Berat badan kamu kurang"
    elif 18.5 <= BMI_format <= 22.9:
        status = "Berat badan kamu normal"
    elif 23.0 <= BMI_format <= 24.9:
        status = "Berat badan kamu berlebih"
    else:
        status = "Kamu mengalami obesitas. Segera lakukan penanganan untuk mencegah timbulnya penyakit"
else:
    status = "Masukan input yang valid"

print("--------------- Hasil Perhitungan BMI ----------------------")

print("Nama:", nama)

print("Jenis kelamin:", jenis_kelamin)

print("Tinggi badan:", tinggi_badan)

print("Berat Badan:", berat_badan)

print("Skor BMI", BMI_format)

print("Status:", nama + ", " + status)
```
